```python
import tensorflow as tf
```

```python
#最急降下法による二次関数の最小化

#パラメータは変数で定義 
x = tf.Variable(0. , name = "x")
#パラメータを使って最小化したい関数を定義
func = (x - 1)**2
```

```python
#これが最急降下法によるパラメータの更新を担当
#learning_rateは一度にずらす大きさ
optimizer = tf.train.GradientDescentOptimizer(
    learning_rate = 0.1
)
#train_stepがxを少しずつずらす操作
train_step = optimizer.minimize(func)
```

```python
#forループでtrain_stepを繰り返す
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(30):
        sess.run(train_step)
    print('x = ', sess.run(x))
```

```python
#機械学習への適用
#Boston house-prices　データセット
```

```python
#データセットの準備
import keras
(x_train, y_train), (x_test, y_test) = keras.datasets.boston_housing.load_data()
```

```python
#matplotlibでデータの確認
% matplotlib inline

import matplotlib.pyplot as plt

plt.rcParams['font.size']  = 10*3
plt.rcParams['figure.figsize'] = [18, 12]
plt.rcParams['font.family'] = ['IPAexGothic']

plt.hist(y_train, bins = 20)
plt.xlabel('住宅価格（＄1,000単位）')
plt.ylabel('データ数')
plt.show()

plt.plot(x_train[:,5],y_train, 'o')
plt.xlabel('部屋数')
plt.ylabel('住宅価格（＄1,000単位）')
```

```python
#データの標準化
x_train_mean = x_train.mean(axis = 0)
x_train_std = x_train.std(axis = 0)
y_train_mean = y_train.mean(axis = 0)
y_train_std = y_train.std(axis = 0)

x_train = (x_train - x_train_mean)/x_train_std
y_train = (y_train - y_train_mean)/y_train_std

#testデータの標準化もtrainデータのstdとmeanを使う
x_test = (x_test - x_train_mean)/x_train_std
y_test = (y_test - y_train_mean)/y_train_std

plt.plot(x_train[:,5], y_train, 'o')
plt.xlabel('number of rooms')
plt.ylabel('prices')
```

```python
#モデルの定義
#ここでは「各説明変数を重み付きで足し合わせたもの」で住宅価格を推定

#説明変数用のplaceholder
#bostonのデータセットは13個の説明変数がある
x = tf.placeholder(tf.float32,(None, 13), name='x')
#正解データ(住宅価格)用のplaceholder
y = tf.placeholder(tf.float32, (None, 1), name = 'y')

#説明変数を重みwで足し合わせただけの簡単なモデル
w = tf.Variable(tf.random_normal((13,1)))
pred = tf.matmul(x,w)

#x は１＊１３、wは13*1だからちょうど掛け合わせられる
```

```python
#損失関数の定義と学習

#誤差の定義とtrain_stepの定義
loss = tf.reduce_mean((y - pred)**2) #reduce_mean()は平均を計算する関数
optimizer = tf.train.GradientDescentOptimizer(
    learning_rate = 0.1
)
train_step = optimizer.minimize(loss)
```

```python
#placeholderのx,yに標準化したx_train, y_trainを代入して
#学習をループさせる

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(100):
        #train_step がNoneを返すので、_で受けておく
        train_loss, _ = sess.run(
            [loss, train_step],
            feed_dict = {
                x: x_train,
                #y_trainとyの次元を揃えるためにreshapeが必要
                y: y_train.reshape((-1,1))
            }
        )
        print('step:{}, train_loss:{}'.format(step, train_loss))
        
    #学習が終わったら、評価用のデータに対して予測してみる
    pred_ = sess.run(
        pred,
        feed_dict={
            x:x_test
        }
    )
```

```python
plt.plot(x_test[:, 5], pred_, 'o', label='prediction')
plt.plot(x_test[:, 5], y_test, 'x', label='realdata')
plt.legend(loc='upper left')
plt.xlabel('num of houses')
plt.ylabel('price')
```

```python
#確率的勾配降下法SGDとミニバッチ学習
#SGDではデータを全て一度の使うのではなく、ミニバッチに分割して学習する

#ミニバッチを1つ処理するのが1イテレーション
#イテレーションを繰り返してデータ全体を処理するのが1エポック
```

```python
#データ全体をシャッフルしてミニバッチに分割し、1つずつ返すジェネレータを作成
import numpy as np

def get_batches(x, y, batch_size):
    n_data  = len(x)
    indices = np.arange(n_data)
    np.random.shuffle(indices)
    x_shuffled = x[indices]
    y_shuffled = y[indices]
    
    #元データからランダムに、batch_size個ずつ抽出
    for i in range(0, n_data, batch_size):
        x_batch = x_shuffled[i : i+batch_size]
        y_batch = y_shuffled[i: i+batch_size]
        yield x_batch, y_batch
```

```python
#ミニバッチを用いた学習

#ミニバッチのサイズ
BATCH_SIZE = 32

step = 0
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    #100エポック回す
    for epoch in range(100):
        for x_batch , y_batch in get_batches(x_train, y_train, 32):
            train_loss, _ = sess.run(
                [loss, train_step],
                feed_dict = {
                    x: x_batch,
                    y: y_batch.reshape((-1,1))
                }
            )
        print('step:{}, train_loss:{}'.format(step, train_loss))
        step += 1
   
    pred_ = sess.run(
        pred,
        feed_dict={
            x: x_test
        }
    )
```

```python
plt.plot(x_test[:, 5], pred_, 'o', label='prediction')
plt.plot(x_test[:, 5], y_test, 'x', label='realdata')
plt.legend(loc='upper left')
plt.xlabel('num of houses')
plt.ylabel('price')
```
