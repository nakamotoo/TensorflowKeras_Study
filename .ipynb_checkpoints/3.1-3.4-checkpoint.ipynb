```python
import tensorflow as tf

a = tf.constant(1, name = "a")
b = tf.constant(1, name = "b")
c = a + b

print(c) #cはテンソル

with tf.Session() as sess:
    print(sess.run(c))　#ここで計算
```

```python
#データフローグラフの表示
graph  =tf.get_default_graph()
print(graph.as_graph_def())
```

```python
#a,bという複数のノードを同時に計算したい時
with tf.Session() as sess:
    sess.run([a,b])
```

```python
#変数->値の更新が可能、学習対象のパラメータなどに使う
import tensorflow as tf

a = tf.Variable(1, name = "a")
b = tf.constant(1, name = "b")

#assignで値を代入し、その結果を返す
c = tf.assign(a, a + b) 

with tf.Session() as sess:
    #変数を利用する場合、sessのはじめに必ず初期化
    sess.run(tf.global_variables_initializer())
    print("1回目[c,a] = ", sess.run([c,a]))
    print("2回目[c,a] = ", sess.run([c,a]))
```

```python
#プレースホルダー
#様々な値を受け取る

a = tf.placeholder(dtype = tf.int32, name = 'a')
b = tf.constant(1, name='b')
c = a+b

with tf.Session() as sess:
    print(sess.run(c, feed_dict={a:1})) #aに１を渡す
```

```python
#演算

a = tf.constant(2, name="a")
b = tf.constant(3, name="b")
c = tf.add(a,b) #a + bと同じ
d = tf.multiply(a, b) #a*bと同じ

with tf.Session() as sess:
    print('a + b = ', sess.run(c))
    print('a + b = ', sess.run(d))
```

```python
#ベクトル演算

a = tf.constant([1,2,3], name = "a")
b = tf.constant([3,4,5], name = "b")

with tf.Session() as sess:
    print(sess.run(a + b))
    print(sess.run(a * b))
```

```python
#行列演算

a = tf.constant([[1,2],[3,4]], name = "a")
b = tf.constant([[1],[2]], name = "b")
#行列の積
c = tf.matmul(a, b) 

print('shape of a: ', a.shape)
print('shape of b: ', b.shape)
print('shape of c: ', c.shape)

with tf.Session() as sess:
    #　\nは改行
    print('a = \n', sess.run(a))
    print('b = \n', sess.run(b))
    print('c = \n', sess.run(c))
```

```python
#テンソル演算とプレースホルダー

#大きさが未知な方向（今回は列方向）はNoneで
a = tf.placeholder(shape = (None, 2), dtype = tf.int32, name = 'a')

with tf.Session() as sess:
    print('--[[1,2]]を代入--')
    print('a = ', sess.run(a, feed_dict= {a: [[1,2]]}))
    print('--[[1,2],[1,2]]を代入--')
    print('a = ', sess.run(a, feed_dict= {a: [[1,2],[1,2]]}))
```

```python
#セッションとSaver
#セッションが変わると、変数は元の値に戻ってしまう

a = tf.Variable(1, name = 'a')
b = tf.assign(a, a+1)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    print('１回目', sess.run(b))
    print('２回目', sess.run(b))
    
with tf.Session() as sess:
    print("新しいセッション")
    sess.run(tf.global_variables_initializer())
    print('１回目', sess.run(b))
    print('２回目', sess.run(b))
```

```python
#Saverによる変数保存

a = tf.Variable(1, name = 'a')
b = tf.assign(a, a+1)

saver = tf.train.Saver()

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    print('１回目', sess.run(b))
    print('２回目', sess.run(b))
    
    #　変数の値をmodel/model.ckptに保存する
    saver.save(sess, 'model/model.ckpt')
    
#Saverを利用する
saver = tf.train.Saver()
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
#変数の値をリストア
    saver.restore(sess, save_path = 'model/model.ckpt')
    print('3回目', sess.run(b))
    print('4回目', sess.run(b))
```
